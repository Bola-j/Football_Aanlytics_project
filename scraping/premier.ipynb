```python
import os
import pandas as pd

def extract_col(col):
    return col[1] if isinstance(col, tuple) else col

def scrape_team(html_file):
    team_name = html_file.replace("_possession.html", "").replace("_", " ")

    with open(html_file, "r", encoding="utf-8") as f:
        html = f.read()

    tables = pd.read_html(html)
    t0 = tables[0]
    t0.columns = [extract_col(c) for c in t0.columns]

    df = t0[["Date","Venue","Opponent","Result","GF","GA","Poss"]].copy()
    df["Team"] = team_name
    df = df[["Team","Date","Venue","Opponent","Result","GF","GA","Poss"]]

    # remove Totals row
    df = df.iloc[:-1, :]
    return df

html_files = [f for f in os.listdir() if f.lower().endswith(".html")]

dfs = []
for f in html_files:
    try:
        d = scrape_team(f)
        dfs.append(d)
        print(f, "rows:", len(d))
    except Exception as e:
        print(f, e)

league_df = pd.concat(dfs, ignore_index=True)
league_df.to_csv("Premeier_League_possession_2023_2024.csv", index=False)

print("DONE rows:", len(league_df), "teams:", league_df["Team"].nunique())
```
